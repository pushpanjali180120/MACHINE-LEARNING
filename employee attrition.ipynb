### We will convert the 2 categorical columns to numerical by label encoding

```python
le = LabelEncoder()
```

```python
k = le.fit_transform(data['salary'])
```

```python
data['salary'] = k
```

```python
data.head()
```

#### Note that we have label encoded the salry column and by default the values 0,1,2 are assigned alphabetically
#### 0 = high, 1 = low, 2 = medium. And also we are dropping the salary column

```python
data.drop(['salary'],inplace = True, axis = 1)
```

```python
data.head()
```

#### Similarly we are encoding the department column and dropping the original column. By default the values will be assigned alphabetically

```python
z = le.fit_transform(data['department'])
```

```python
data['department'] = z
```

```python
data.head()
```

```python
data.drop(['department'], inplace = True, axis = 1)
```

```python
data.dtypes ## all the categorical columns are converted to numerical.
```

### Splitting the datasets to independent and target variables

```python
get_ipython().system('pip install imblearn')
```

### Since we know that the target column is highly imbalanced, we will use SMOTE to balance the train data to avoid overfitting

```python
from imblearn.over_sampling import SMOTE
from sklearn.model_selection import train_test_split
```

```python
# Separate input features and target
x = data.drop(['left'],axis=1)
y = data['left']
# setting up testing and training sets
x_train,x_test,y_train,y_test = train_test_split(x,y, test_size = 0.3, random_state=0)
sm = SMOTE(random_state=27)
x_train, y_train = sm.fit_resample(x_train, y_train)
```

```python
y_train = pd.DataFrame(y_train, columns = ['left'])
y_train.left.value_counts()
```

### Model building

### 1. Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dt = DecisionTreeClassifier()
```

```python
dt.fit(x_train,y_train)
```

```python
prediction_dt = dt.predict(x_test)
```

```python
prediction_dt
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, recall_score,precision_score,roc_curve,auc
```

```python
accuracy_dt = accuracy_score(y_test, prediction_dt)*100
accuracy_dt
```

```python
print(classification_report(y_test, prediction_dt))
```

```python
print(confusion_matrix(y_test, prediction_dt))
```

```python
feature_importance = pd.DataFrame(dt.feature_importances_,index=x_train.columns,columns=['Importance']).sort_values('Importance',ascending=False)
```

```python
feature_importance
```

## 2.KNN

#### We will first go ahed and standardize the data before buliding further models using standard scaler

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler().fit(x_train)
x_train_std = sc.transform(x_train)
x_test_std = sc.transform(x_test)
```

```python
x_train_std
```

```python
x_test_std
```

```python
from sklearn.neighbors import KNeighborsClassifier
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(x_train_std, y_train)
```

```python
prediction_knn = knn.predict(x_test_std)
```

```python
accuracy_knn = accuracy_score(y_test, prediction_knn)*100
```

```python
accuracy_knn
```

```python
print(classification_report(y_test, prediction_knn))
```

```python
confusion_matrix(y_test, prediction_knn)
```

```python
algorithms =['Decision Tree', 'KNN']
scores=[accuracy_dt, accuracy_knn]
sns.barplot(x=algorithms,y=scores)
plt.xlabel('Algorithms')
plt.ylabel('Accuracy score')
plt.show()
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf = RandomForestClassifier()
```

```python
rf.fit(x_train,y_train)
```

```python
prediction_rf = rf.predict(x_test)
```

```python
prediction_rf
```

```python
accuracy_rf = accuracy_score(y_test, prediction_rf)*100
accuracy_rf
```

```python
print(classification_report(y_test, prediction_rf))
```

```python
print(confusion_matrix(y_test, prediction_rf))
```

```python
compare = pd.DataFrame({'Model': [ ' Decision Tree train',' KNN',' Random Forest',], 'Accuracy': [accuracy_dt,accuracy_knn, accuracy_rf] })
compare.sort_values(by='Accuracy', ascending=False)
```

```python
plt.style.use("seaborn")

x = [ ' Decision Tree train',' KNN',' Random Forest']

y = [accuracy_dt,accuracy_knn, accuracy_rf]

fig, ax = plt.subplots(figsize=(5,5))
sns.barplot(x=x,y=y, palette="hot");
plt.ylabel("Model Accuracy")
plt.xticks(rotation=40)
plt.title("Model Comparison - Model Accuracy data", fontsize=15, fontname="Helvetica", y=1.03);
```
